```python
df_drinks['beer_volume_per_serving_ltr'] = .35
df_drinks['spirit_volume_per_serving_ltr'] = .04
df_drinks['wine_volume_per_serving_ltr'] = .15
```

```python
df_drinks.head()
```

```python
df_drinks['total_litres_of_pure_alcohol'] = (df_drinks['beer_servings'] * df_drinks['beer_ABV'] * df_drinks['beer_volume_per_serving_ltr']) + (df_drinks['spirit_servings'] * df_drinks['spirit_ABV'] * df_drinks['spirit_volume_per_serving_ltr']) + (df_drinks['wine_servings'] * df_drinks['wine_ABV'] * df_drinks['wine_volume_per_serving_ltr']) 
```
